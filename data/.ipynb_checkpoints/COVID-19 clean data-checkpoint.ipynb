```python
import pandas as pd
import numpy as py
```

```python
cov19 = pd.read_csv("covid_19_clean_complete.csv")
cov19.head()
```

```python
cov19 = cov19.drop(columns = ['Province/State','Lat','Long','WHO Region'])
```

```python
cov19 = cov19.rename(columns = {'Country/Region':'Country'})
cov19.head()
```

```python
#group data
country_count = cov19[cov19.Date == cov19.Date.max()].groupby('Country')['Confirmed',
                                                                     'Deaths',
                                                                     'Recovered','Active'].sum().sort_values(by='Confirmed',ascending=False)
```

```python
country_count.head()
```

```python
#save csv
country_count.to_csv('cov19_country_count.csv')
```

```python
date_clean = cov19.groupby('Date')['Confirmed','Deaths','Recovered'].sum()
date_country_count = cov19.groupby(['Date','Country'])['Confirmed','Deaths','Recovered'].max()
date_country_count.head()
```

```python
#save csv
date_country_count.to_csv('cov19_date_country_count.csv')
```
