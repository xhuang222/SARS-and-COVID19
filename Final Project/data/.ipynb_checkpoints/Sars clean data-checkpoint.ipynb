```python
import pandas as pd
import numpy as py
```

```python
sars = pd.read_csv("sars_2003_complete_dataset_clean.csv")
```

```python
sars.head()
```

```python
sars = sars.rename(columns = {'Cumulative number of case(s)':'Confirmed',
                             'Number of deaths':'Deaths',
                             'Number recovered':'Recovered'})
```

```python
sars.head()
```

```python
sars.Date = pd.to_datetime(sars.Date)
```

```python
#group data
country_count = sars[sars.Date == sars.Date.max()].groupby('Country')['Confirmed',
                                                                     'Deaths',
                                                                     'Recovered'].sum().sort_values(by='Confirmed',ascending=False)
```

```python
country_count.head()
```

```python
country_count['Confirmed']=country_count['Confirmed'].apply(str)
country_count['Deaths']=country_count['Deaths'].apply(str)
country_count['Recovered']=country_count['Recovered'].apply(str)
```

```python
country_count.dtypes
```

```python
#save csv
country_count.to_csv('sars_country_count.csv')
```

```python
#group data
date_clean = sars.groupby('Date')['Confirmed','Deaths','Recovered'].sum()
```

```python
date_clean.head()
date_clean.to_csv('date_count.csv')
```

```python
date_country_count = sars.groupby(['Date','Country'])['Confirmed','Deaths','Recovered'].max()
date_country_count.head()
```

```python
#save csv
date_country_count.to_csv('sars_date_country_count.csv')
```

```python
sars_cases = sars.pivot_table(index='Country', columns = 'Date', values='Confirmed')
sars_cases.head()
```

```python
sars_cases.to_csv('sars_timeseries.csv')
```
